### Question 2
<h4>What does the distribution of customer rating looks like? Draw the mean,25th and 75th percentile on the plot.<br><br>
    Write the python code in the below cell to create the appropriate visual to perform the above task .
<h4>Answer in markdown cells below the visual <br><br>
    1.Summarise your findings from the visual.<br>

```python
fig, ax = plt.subplots(figsize=(12, 4))
sns.boxplot(data=df, x='Rating', ax=ax, width=0.5)

# Calculate mean, 25th, and 75th percentiles
mean_rating = df['Rating'].mean()
percentile_25 = df['Rating'].quantile(0.25)
percentile_75 = df['Rating'].quantile(0.75)

# Plot mean, 25th, and 75th percentiles as vertical lines
plt.axvline(mean_rating, color='red', linestyle='--', label='Mean')
plt.axvline(percentile_25, color='green', linestyle='--', label='25th Percentile')
plt.axvline(percentile_75, color='orange', linestyle='--', label='75th Percentile')
plt.legend()
plt.title("Distribution of Customer Rating with mean, 25 and 75 Quartile.")

plt.show()
```

1. **Findings from the Visual:**
   - The boxplot illustrates the distribution of customer ratings, with key summary statistics marked by horizontal lines within the box.
   - The median rating (50th percentile) appears to be around 7, as indicated by the red dashed line.
   - The interquartile range (IQR), represented by the width of the box, suggests that the majority of ratings fall between approximately 5.5 and 8.5 
   - Outliers are not present in this box plot.

2. **Reason for Selecting Boxplot:**
   - We choose a boxplot because it effectively summarizes the distribution of numerical data and highlights key summary statistics such as the median, quartiles, and outliers.
   - It allows for easy comparison of the rating distribution and provides insights into the spread and central tendency of the data.

3. **Pre-attentive Attributes Used:**
   - Length: The length of the box and whiskers provides a quick understanding of the spread of the ratings, with longer whiskers indicating a larger range of ratings.
   - Color: Different colors are used for the mean and quartile lines, making them stand out from the rest of the plot and drawing attention to these key summary statistics.

4. **Gestalt Principles Used:**
   - Proximity: The proximity of the quartile lines and the mean line to the boxplot elements suggests their association with the distribution of ratings, aiding in the interpretation of summary statistics.
   - Closure: The closed box in the boxplot visually encapsulates the interquartile range, making it easy to identify and compare the central spread of ratings.

## Comparision of Product lines based on Gender.

```python
# Create a stacked bar plot
plt.figure(figsize=(14, 12))
#sns.palplot(sns.color_palette("BuGn_r", 10)) 
sns.countplot(data=df, x='Gender', hue='Product line', palette='viridis')

# Add labels and title
plt.xlabel('Gender Category')
plt.ylabel('Count')
plt.title('Comparison of Product Lines based on Gender')

# Rotate x-axis labels for better readability
plt.xticks()

# Show plot
plt.legend(title='Product Line')
plt.tight_layout()
plt.show()
```

**Findings from the visual:**
- The visual presents a comparison of product lines based on gender categories (Male, Female). Each bar represents the count of different product lines for each gender category.
- In the *Female* category *Fashon accessories*, *Food and Beverages* are the highest prefered product lines.
- In the *Male* category *Health and beauty* and *Electronic accessories* are the highest prefered product lines.

**Reason for selecting the chart type:**
- A grouped bar chart is chosen because it allows for the comparison of multiple categories (product lines) within each gender category.
- Grouped bars visually represent the composition of each category (gender) in terms of subcategories (product lines), making it easy to compare the distribution of product lines across genders.

**Pre-attentive attributes used:**
- **Color:** Different colors are used to represent different product lines within each gender category. This helps in distinguishing between the product lines quickly.
- **Length/Height:** The length of each stacked bar represents the count of product lines, making it easy to compare the counts visually.

**Gestalt principles used:**
- **Proximity:** Within each gender category, the product lines are grouped closely together, emphasizing their relationship and making it easier for viewers to associate each product line with its respective gender category.
- **Similarity:** Product lines belonging to the same gender category are visually similar in terms of color (belonging to the same palette) and shape (grouped bars), aiding in the perception of groups and facilitating comparison within each gender category.
